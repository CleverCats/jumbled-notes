```python
import matplotlib

# 默认无衬线字体,设置宋体
# matplotlib.rcParams['font.sans-serif'] = ['SimHei']

# plot规则 marker|line|color
x5 = np.array([1, 3, 2, 6])
y5 = np.array([2, 4, 6, 8])
# 格式化规则（推荐）：o:r 点|虚线|红色
# ms[marker size] 标记点的尺寸
# mec[marker edge color] 点的边缘颜色(140中英文全称or首字母)
# mfc[marker face color] 点的表面颜色(140中英文全称or首字母)
# ls[line style] 线型solid/dotted/dashed/（dashed/dotted）
# color/c 线条颜色
# lw[linw width] 线宽

# 两次调用plot或者plt.plot(x1,y1, x2, y2)实现多线绘制
plt.plot(x5, y5,'o-.r',ms=15,mec='g', mfc='pink',lw=1)
plt.plot(x5-2, y5,'D-g',ms=5,mec='b', mfc='r',lw=2)


# 标题[上]/x轴[下]/y轴[左]
# fontdict[字体属性字典]  dict{字体/颜色/尺寸}
# loc[位置] 默认center
# xlabel和标题：left/center/right ylabel
# ylabel： bottom/center/top
front1 = {'family':'SimHei','color':'blue','size':15}
plt.title('重量体积变化', fontdict=front1)
plt.xlabel('物体重量', color='green',loc='right')
plt.ylabel('物体体积',color='r',loc='top')

# 设置网格线 默认x/y轴都有
plt.grid(color='b', axis='y' ,ls='-', lw=0.5)
plt.show()
```

```python
# 多图绘制
x6,y6=   np.array([0, 1, 2, 3]), np.array([2, 5, 8, 7])
x7,y7=   np.array([0, 1, 2, 3]), np.array([1, 3, 4, 5])
x8,y8=   np.array([0, 1, 2, 3]), np.array([0, 5, 2, 3])
x9,y9=   np.array([0, 1, 2, 3]), np.array([0, 1, 5, 3])

# subplot(行规格,列规格,索引) 2*2=4 index 1->4
plt.subplot(2, 2, 1)
plt.title('1',color='g')
plt.plot(x6,y6)

plt.subplot(2, 2, 2)
plt.title('2',color='g')
plt.plot(x7,y7)

plt.subplot(2, 2, 3)
plt.title('3',color='g')
plt.plot(x8,y8)

plt.subplot(2, 2, 4)
plt.title('4',color='r')
plt.plot(x9,y8)

# 大标题
plt.suptitle('折线图',family='SimHei')
plt.show()
```

```python
# 散点[scatter]图 
# alpha 透明度
xx1 = np.array(range(1,10))
yy1 = np.array(range(5,14))
plt.scatter(xx1, yy1,c='g',alpha=0.2,s=10)

xx2 = np.array(range(25,45))
yy2 = np.array(range(40,60))
plt.scatter(xx2, yy2,c='r')


xx3 = np.array([10,20,30,40])
yy3 = np.array([60,70,80,90])

# 每个点上色传入颜色数组必须使用参数c len(colors)==len(xx3)
colors =np.array(['red','green','blue','yellow'])
plt.scatter(xx3, yy3,c=colors)
plt.show()
```

```python
# 颜色图
plt.scatter(xx1,yy1)
plt.scatter(xx2,yy2)
plt.scatter(xx3,yy3)
plt.colorbar()
```

```python
# 柱状图[适用分类数据]  x轴表类别,y轴表大小
xx4 = np.array(['A','B','C'])
yy4 = np.array([2, 5, 8])
# 垂直柱状图 bar width
plt.subplot(1,2,1)
colors = np.array(['g','r','y'])
plt.bar(xx4,yy4,width=0.1,color=colors)

# 水平柱状图 barh height
plt.subplot(1,2,2)
plt.barh(xx4,yy4,height=0.1)
plt.show()
```

```python
# hist 直方图[适合展示数据分布] x轴表标量大小,y轴表频次
import torch
datas = torch.normal(5,10,(1, 30))
print(datas)
plt.hist(datas)
plt.xlabel('标量',c='g',family='SimHei',size=20)
plt.ylabel('频次',c='g',family='SimHei',size=20)

plt.show()
```

```python
# 饼状图
datas = np.array([10,20,30,40])
# 标签
lables = ['A','B','C','D']
# 距离圆心的距离（默认0）[突出显示]
explode = [0, 0.1, 0, 0]
# shadow 阴影设置[增加立体感]
plt.pie(datas, labels=lables, explode=explode, shadow=True)
# 图例标签
plt.legend(title='TL')
plt.show()
```
